```python
# train,test and split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
# As sklearn does not accept categorical data we need to convert the sentences to numbers using countvectoriser
from sklearn.feature_extraction.text import CountVectorizer 
vect = CountVectorizer(stop_words='english')
x_train_vect = vect.fit_transform(x_train)
x_test_vect = vect.transform(x_test)
```

```python
count=0
for i in range(len(x_train)):
    if 'yummy' in x_train[i]:
        count+=1
print(count)
```

```python
temp=vect.get_feature_names_out()
temp
```

```python
vect.vocabulary_
```

```python
x_train_vect.toarray() #gives a sparse representation of numbers after converting from text 
```

```python
# Using logistic regression
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train_vect,y_train)
y_pred=model.predict(x_test_vect)
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
# Building the model
from sklearn.svm import SVC
```

```python
# First model with only SVC
model1=SVC()
```

```python
model1.fit(x_train_vect,y_train)
```

```python
y_pred1=model1.predict(x_test_vect)
y_pred1
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred1)
```

```python
# model.predict(["Wow... Loved this place."]) # it does not accept text data, we have convert it to numerical data using count vectorizer 
```

```python
# Model2 using pipeline model
# Pipeline model is combination of SVC + CountVectoriser
from sklearn.pipeline import make_pipeline 
model2 = make_pipeline(CountVectorizer(),SVC())
model2.fit(x_train,y_train) 
y_pred2 = model2.predict(x_test)
y_pred2
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred2)
```

```python
# Model3 using NAIVEBAYES
from sklearn.naive_bayes import MultinomialNB
```

```python
# Building the model
model3=MultinomialNB()
```

```python
model3.fit(x_train_vect,y_train)
```

```python
y_pred3=model3.predict(x_test_vect)
y_pred3
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred3)
```

```python
# Model4 using NAIVEBAYES and CountVectoriser
# PIPELINE WITH NAIVENAYES
from sklearn.pipeline import make_pipeline 
model4 = make_pipeline(CountVectorizer(),MultinomialNB())
model4.fit(x_train,y_train) 
y_pred4 = model4.predict(x_test)
y_pred4
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred4)
```

```python
# Accuracy scores of all models
# model1 - 0.72  (Support Vector Classifier)
# model2 - 0.792 (SVC + CountVectorisor)
# mode13 - 0.744 (Naive Bayes)
# mode14 - 0.784 (Naive Bayes+ CountVectorisor)
```
